```python
import pandas as pd
```

```python
mvps = pd.read_csv("mvps.csv",encoding="utf-8")
```

```python
mvps
```

```python
mvps = mvps[["Player","Year","Pts Won", "Pts Max","Share"]]
```

```python
mvps.head()
```

```python
players = pd.read_csv("players.csv",encoding="utf-8")
```

```python
players
```

```python
del players ["Unnamed: 0"]
del players ["Rk"]
```

```python
players.head()
```

```python
# bazi player isimlerinde * var onlari kaldiriyoruz (galiba hall of fame veya mvp yarisinda olanlar)
players["Player"] = players["Player"].str.replace("*","",regex=False)
```

```python
players.head(50)
```

```python
# bazi transfer olup birden fazla olan oyuncular birden fazla yazilmis bize sadece total laazim
players.groupby(["Player","Year"])
```

```python
def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Team"] == "TOT"] #tot total statsini veren satir butun takimlardaki eger bir yil icinde  birden fazla takim degistirdiyse
        row["Team"] = df.iloc[-1,:]["Team"]
        return row

players = players.groupby(["Player","Year"]).apply(single_row)
```

```python
players.head(50)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
combined = players.merge(mvps,how = "outer", on= ["Player","Year"])
```

```python
combined.head(5) #NaN olanlar mvp olamyanlar onlari fixleyecegiz
```

```python
combined[["Pts Max" , "Pts Won" , "Share"]].fillna(0,inplace = True) #copy aldigimiz dataset ile inplace yapiyoruz bunun yerine assign yaparsak bu hatayi almayiz
```

```python
combined[["Pts Max" , "Pts Won" , "Share"]] = combined[["Pts Max" , "Pts Won" , "Share"]].fillna(0)
```

```python
combined
```

```python
players
```

```python
combined["Player"] = combined["Player"].replace({
    "Nikola JokiÄ": "Nikola Jokić",
    "Nikola JokiÄ‡": "Nikola Jokić",
    "Luka DonÄiÄ": "Luka Dončić"
})
```

```python
combined
```

```python
teams = pd.read_csv("teams.csv")
```

```python
teams
```

```python
teams.head(30)
```

```python
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams.head(30)
```

```python
teams["Team"] = teams["Team"].str.replace("*","",regex=False)
```

```python
teams.head(5)
```

```python
teams["Team"].unique()
```

```python
combined["Team"].unique()
```

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]: #skip the first row
        abbrev,name = line.replace("\n","").split(",")
        nicknames[abbrev] = name
```

```python
nicknames
```

```python
combined["Team"] = combined["Team"].map(nicknames)
```

```python
combined.head(5)
```

```python
stats = combined.merge(teams,how="outer", on = ["Team","Year"])
```

```python
stats
```

```python
del stats["Unnamed: 0"]
```

```python
stats.dtypes
```

```python
stats = stats.apply(pd.to_numeric, errors = "ignore")
```

```python
stats.dtypes
```

```python
stats["GB"].unique()
```

```python
stats["GB"] = stats["GB"].str.replace("—","0")
```

```python
stats["GB"].unique()
```

```python
stats["GB"] = pd.to_numeric(stats["GB"])
```

```python
stats.dtypes
```

```python
stats.to_csv("player_mvp_stats.csv")
```
